```python
#Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Importing the data
dataset = pd.read_csv(r'C:\Users\ayushsingh\OneDrive - Sopra Steria\Desktop\Machine Learning\Data Analyst\Finance_data.csv')
```

```python
#Checking for the datasets columns
dataset.head()
```

```python
#Getting the info of datasets whether any columns have any null data and dataset column Dtype
dataset.info()
```

```python
#Plotting the boxplot to check which gender do the investements
sns.boxplot(x="Gender",y="Age",data=dataset)
plt.show()
```

```python
#Dropping the null values
dataset.dropna()
```

```python
#Dropping the Equilty Market Column
dataset.drop(["Equity_Market"],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
#dropping the columns
dataset.drop(["Debentures","Government_Bonds"],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
#dropping the columns
dataset.drop(["Fixed_Deposits","PPF","Gold"],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
#Plotting the investment_avenue and stock market graph 
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.countplot(x=dataset["Gender"],hue=dataset["Investment_Avenues"]) # Single value can be passed in counterplot x,y
plt.title("Investment_Avenues")
plt.show()


plt.figure(figsize=(12,6))
plt.subplot(1,2,2)
sns.countplot(x=dataset["Gender"],hue=dataset["Stock_Market"])
plt.title("Stock Market")
plt.show()
```

```python
plt.figure(figsize= (18,6))
plt.subplot(1,2,1)
sns.countplot(hue = dataset['Gender'], x = dataset['Duration'], palette = 'viridis', linewidth = 2, edgecolor = "black" )
plt.title("DURATION")

plt.subplot(1,2,2)
sns.countplot(hue = dataset["Gender"], x = dataset["Invest_Monitor"], palette = "seismic", linewidth = 2, edgecolor ="black")
plt.title("INVESTMENT MONITORING")
plt.show()
```
